**PLOT 1**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

fig,axes = plt.subplots(1,3,figsize = (15,7))

x = np.linspace(0,4,50)
x1, x2 = np.cos(x),np.sin(x)

axes[0].set_title('plot 1')
axes[0].loglog(x,x1,"g",x,x2,'r')

axes[1].set_title("plot 2")
axes[1].semilogy(x,x1,'g',x,x2,'r')

axes[2].set_title("plot 3")
axes[2].plot(x,x1,'g',x,x2,'r')
axes[2].set_xscale("log")
axes[2].set_yscale("log")
```

**loglog():** plots on a logarithmic scale

**semilogy()**: sets y-axis to a logarithmic scale, while x-axis is linear

**set_xscale(), set_yscale()**: x-axis and y-axis are manually set to logarithmic scales.

**PLOT-2**

```python

fig, ax = plt.subplots(1,2,figsize=(12,4))

h = 5
x = np.linspace(0,10,50)
S_A = (2*np.pi*x**2)+(2*np.pi*(x**2)*h)
V = np.pi*(x**2)*h

ax[0].set_title("Surface area of cylinder")

ax[0].plot(x,S_A,lw =2, color="green")
ax[0].set_ylabel(r"surface area($m^2$)",fontsize = 16, color="red")
for label in ax[0].get_yticklabels():
  label.set_color("red")

ax[1].set_title("Volume of cylinder")
ax[1].plot(x,V,lw=2,color="red")
ax[1].set_ylabel(r"Volume($m^3$)",fontsize=19,color="red")
for label in ax[1].get_yticklabels():
  label.set_color("red")
```

**get_yticklabels()**: Returns a list of ticks on y-axis

**PLOT-3**

```python
fig,axes = plt.subplots(1,2,figsize=(12,6),sharex = True, sharey = True)

x1 = np.random.randn(500)
x2 = np.random.randn(500)

axes[0].set_title("uncorrelated")
axes[0].scatter(x1,x2)

axes[1].set_title("positive correlation")
axes[1].scatter(x1,x1+0.4*x2)
```
