```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

x = np.linspace(-3*np.pi,3*np.pi,100)
y = np.cos(2*np.pi*x)+np.sin(2*np.pi*x)

fig,axes = plt.subplots(1,4,figsize = (20,7))

axes[0].plot(x,y,lw=3)
axes[0].set_title("default_ticks")

axes[1].plot(x,y,lw=3)
axes[1].set_title("set_xticks")
axes[1].set_yticks([-1,0,1])
axes[1].set_xticks([-5,0,5])

axes[2].plot(x,y,lw=3)
axes[2].set_title("setmajor_locator")
axes[2].xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
axes[2].yaxis.set_major_locator(mpl.ticker.FixedLocator([-1,0,1]))
axes[2].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))


axes[3].plot(x,y,lw=2)
axes[3].set_title("set_xticklabels")
axes[3].set_yticks([-1,0,1])
axes[3].set_xticks([-3*np.pi,-2*np.pi,-np.pi,0,np.pi,2*np.pi,3*np.pi])
axes[3].set_xticklabels([r'$-3\pi$',r'$-2\pi$',r'$-\pi$',0,r'$\pi$',r'$2\pi$',r'$3\pi$'])

x_minor_ticker = mpl.ticker.FixedLocator([-2*np.pi,-1*np.pi,0,np.pi,2*np.pi])

axes[3].xaxis.set_minor_locator(x_minor_ticker)
axes[3].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(4))
```

**Ticks**: Ticks are used in plots to mark specific values on the axes.

**axes[0].set_title("default_ticks")**: This function sets a title to the subplots. Here, no customization is given so the matplotlib function gives default setttings.

**axes[1].set_yticks([-1,0,1])**

**axes[1].set_xticks([-5,0,5])**: This function gives specification on where to place the ticks. For Y-axis ticks are placed at -1, 0 and 1. For x-axis ticks are placed at -1,0, and 1.

**axes[2].xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))**: The major x-ticks have a maximum of 4 ticks, which is spaced evenly.

**axes[2].yaxis.set_major_locator(mpl.ticker.FixedLocator([-1,0,1]))**: Fixes the y-axis at -1,0 and 1.

**axes[2].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))**: sets a maximum of 8 ticks.

**axes[3].xaxis.set_minor_locator(x_minor_ticker)**: places minor ticks in between the major ticks.
**axes[3].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(4))**:  Gives the maximum number of minor ticks  that can be placed in y-axis.Here, it is 4.
