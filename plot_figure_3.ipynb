**PLOT 1**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd

x = np.linspace(-10,10,500)
y = np.sin(x) + np.cos(x)

fig, ax = plt.subplots(figsize=(10,4))

ax.plot(x,y, linewidth = 2)

ax.spines["right"].set_color('blue')
ax.spines['top'].set_color('blue')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

ax.set_xticks([-10,-5,5,10])
ax.set_yticks([0.5,1])

for label in ax.get_xticklabels() + ax.get_yticklabels():
  label.set_bbox({'facecolor':'white','edgecolor':'blue'})
```

**ax.spines["right"].set_color('blue')**

**ax.spines['top'].set_color('blue')**: The right and top spines are blue

**set_ticks_position**: To emphasize specific values  both x and y axis have user defined ticks.

**set_position**: is used to position spines of a plot.

**for label in ax.get_xticklabels() + ax.get_yticklabels():

  label.set_bbox({'facecolor':'white','edgecolor':'blue'})**: The first part combines the tick labels of both x-axis and y-axis. Iterates through each tick label  and customizes the bounding box.

**PLOT - 2**

```python
fig = plt.figure(figsize = (10,4))

def f(x):


  return (1/(x**2))*(np.cos(x))

def plot(ax,x,f,fontsize):
  ax.plot(x,f(x),linewidth = 5)

  ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
  ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(3))
  ax.set_xlabel(r"$x$",fontsize=fontsize)

  ax.set_ylabel(r"f(x)", fontsize=fontsize)

ax = fig.add_axes([0.1,0.15,0.8,0.9],facecolor="#f5f5f5")
x = np.linspace(0,14,100)

plot(ax,x,f,12)

x0,x1 = 5,7
ax.axvline(x0,ymax = 0.7,color="blue",linestyle="-")
ax.axvline(x1,ymax=0.7,color="blue",linestyle=":")

ax1 = fig.add_axes([0.6,0.5,0.3,0.5],facecolor="lightblue")
x = np.linspace(x0,x1,50)
plot(ax1,x,f,10)
```


**ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))**:  This code makes sure that axes have 5  evenly spaced ticks

**ax.axvline(x0,ymax = 0.7,color="blue",linestyle="-")**: Draws a vertical line at 0.7

**PLOT-3**

```python
x = y = np.linspace(-10,10,150)

X,Y = np.meshgrid(x,y)

Z = np.cos(X)*np.exp(-(Y/5)**2)

fig, ax = plt.subplots(figsize=(7,5))

norm = mpl.colors.Normalize(-abs(Z).max(),abs(Z).max())
p = ax.pcolor(X,Y,Z,norm = norm, cmap = mpl.cm.Blues)

ax.axis('tight')
ax.set_xlabel(r"$x$",fontsize=9)
ax.set_ylabel(r"$y$",fontsize=9)

ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(9))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(9))

cb = fig.colorbar(p, ax=ax)
cb.set_label(r"$z$", fontsize = 18)
cb.set_ticks([-1,-0.5,0,0.5,1])
```

**mpl.colors.Normalize(-abs(Z).max(),abs(Z).max())**: Used to normalize the data values to the range [0,1]. abs.max() makes sure that normalization is symmetric around zero.

**ax.pcolor(X,Y,Z,norm = norm, cmap = mpl.cm.Blues)**: norm applies normalization and creates a pseudocolor plot.
