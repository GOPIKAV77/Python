Seaborn is used for statistical plotting in python.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("TESLA.csv")
```

```python
df.head()
```

```python
sns.lineplot(x="Open",y="Close",data = df,color="red")
```

```python
sns.lineplot(x="Open",y="High",data=df)
```

```python
sns.lineplot(x="Open",y="Volume",data=df)
sns.set_style("darkgrid")
sns.set_context("talk")
sns.despine()
plt.show()
```

```python
sns.stripplot(x="Open",y="High",data=df)
```

```python
sns.stripplot(x="Open",y="Close",data=df,jitter=True,color="red")
```

```python
sns.displot(df["Volume"])
```

```python
sns.lmplot(x="Open",y="Close",data=df)
```
