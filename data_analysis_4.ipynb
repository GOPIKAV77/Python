```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("cars.csv")
```

```python
d = df.drop("disp",axis=1).groupby("mpg").sum().sort_values("cyl",ascending=False)
```

```python
c = df.mpg.value_counts()
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,4))
c.plot(kind='hist',ax=ax1)
d.cyl.plot(kind="hist",ax=ax2)
```
