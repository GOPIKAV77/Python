```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

s = sp.symbols("S")
Sk = 4
s1 = sp.cos(s**2)-2*sp.sin(s)

f = lambda S: sp.lambdify(s,s1,'numpy')(S)
fp = lambda S: sp.lambdify(s, sp.diff(s1,s),'numpy')(S)

S = np.linspace(1,5,100)

fig,ax = plt.subplots(1,1,figsize=(12,4))
ax.plot(S,f(S))
ax.axhline(0,ls=":",color='r')


n = 0
while f(Sk) > 1:
  Sk_new = Sk - f(Sk)/ fp(Sk)

  ax.plot([Sk, Sk],[0, f(Sk)],color='r',ls=":")
  ax.plot(Sk,f(Sk),'k')
  ax.text(Sk,-1,r"$x_%d$"%n,ha='center')

  ax.plot([Sk,Sk_new],[f(Sk),0],'k')

  Sk = Sk_new

  n += 1

ax.plot(Sk,f(Sk),'r*',markersize = 20)

ax.annotate("Root is at %.3f"%Sk, fontsize = 20, family = "serif", xy = (Sk,f(Sk)),xycoords = 'data',xytext=(-150,150),textcoords = 'offset points', arrowprops = dict(arrowstyle ="->",connectionstyle="arc3, rad = .5"))

ax.set_xticks([-1])
```
