Regression Equation: Can be used to find out relationship between variables. It is used to find out outcomes of future events.


```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
from scipy import stats
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel("icecream.xlsx")
df
```

```python
plt.scatter(df["Student_Population"],df["Sales"],color="red")
plt.ylabel("sales")
plt.xlabel("student_population")
```

```python
import statsmodels.api as s
population = df["Student_Population"]
sale = df["Sales"]
population = s.add_constant(population)
model = s.OLS(sale, population)
result = model.fit()
result.summary()
```

```python
x = df["Student_Population"].values.reshape(-1,1)

y = df["Sales"].values.reshape(-1,1)
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_[0],reg.coef_[0][0]
```

Hence, the estimated regression equation is, y = 60+sx

```python
plt.scatter(df["Student_Population"],df["Sales"],color="red")
plt.plot([min(df["Student_Population"]),max(df["Student_Population"])],[reg.intercept_[0],reg.coef_[0][0]*max(df["Student_Population"])+reg.intercept_[0]],color="black")
plt.text(15,120,"y=60+5x")
```
