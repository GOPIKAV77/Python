```python
import pandas as pd
data = pd.read_csv("tips.csv")
```

```python
data.head() #Gives first few rows
```

```python
# To delete a particular row use drop
data = data.drop(4)
data
```

```python
#scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill",y="tip",data=data)
plt.show()
```

```python
#for distinguishing smokers
import seaborn as sns
sns.scatterplot(x="total_bill",y="tip",hue="smoker",data=data)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x="total_bill",y="tip",hue="sex",data=data)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x="total_bill",y="size",hue="sex",data=data)
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x="total_bill",y="size",data=data)
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x="total_bill",y="smoker",data=data)
plt.show()
```
