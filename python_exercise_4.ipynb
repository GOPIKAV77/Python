```python
#Get a word from user and print all possible combinations of alphabets in that word
```

```python

def swap(string,i,j):
    temp = list(string)
    temp[i], temp[j] = temp[j], temp[i]
    return''.join(temp)

def generate_combinations(string, start, end):
  if start == end:
    print(string)

  else:
    for i in range(start, end+1):
      string =  swap(string, start, i)

      generate_combinations(string, start + 1, end)
      string = swap(string, start , i)



print("Give a word to print all possible combinations :")
word = input()

print("All possible combinations are :")
generate_combinations(word,0,len(word)-1)
```
