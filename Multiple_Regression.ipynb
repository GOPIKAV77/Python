Multiple Regression Model

```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("TRUCKING.xlsx")
df
```

Scatter diagram of data for x1

```python
import matplotlib.pyplot as plt
plt.scatter(df["x1"],df["travel_time"],color="green")
plt.ylabel('Travel time')
plt.title('Linear Regression for x1')
```

Scatter diagram for no of deliveries

```python
plt.scatter(df["n_of_deliveries"],df["travel_time"],color="red")
plt.ylabel("Travel time")
plt.title('linear regression for number of deliveries')
```

scatter diagram for x1 and number of deliveries

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(df["x1"],df["travel_time"],color="green")
plt.scatter(df["n_of_deliveries"],df["travel_time"],color="blue")
plt.ylabel("Travel time")
plt.title("Multiple Regression")
```

Linear Regression Vs Multiple Regression

Linear Regression

```python
reg = ols(formula="travel_time~x1",data = df)
fit = reg.fit()
fit.summary()
```

Multiple Linear Regression

```python
from statsmodels.formula.api import ols
model = ols("travel_time~x1+n_of_deliveries",data = df).fit()
model.summary()
```

The equation is y = -0.8687+0.0611 x1 + 0.9234 x2

Anova

```python
print(anova_lm(fit))
```

```python
anova_table = anova_lm(model,typ=1)
anova_table
```

OLS summary

```python
from statsmodels.formula.api import ols
model = ols("travel_time~x1+n_of_deliveries",data=df).fit()
model.summary()
```
