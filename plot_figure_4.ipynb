**PLOT**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl

fig,axes = plt.subplots(1,3,figsize=(14,4),subplot_kw = {'projection':'3d'})

def title(ax,title):
  ax.set_title("plot")
  ax.set_xlabel("$x$",fontsize=14)
  ax.set_ylabel("$y$",fontsize=14)
  ax.set_zlabel("$z$",fontsize=14)

x = y = np.linspace(-3,3,74)
X,Y = np.meshgrid(x,y)

S = np.sqrt(X**2)
Z = np.cos(S*Y)

norm = mpl.colors.Normalize(-abs(Z).max(),abs(Z).max())

P = axes[0].plot_surface(X,Y,Z,rstride = 2, cstride = 2, linewidth = 1, antialiased = True, norm=norm,cmap = mpl.cm.magma)

cb = fig.colorbar(P,ax=axes[0],shrink = 1)

title(axes[0],"plot_surface")

P = axes[1].plot_wireframe(X,Y,Z,rstride=2,cstride=2,color="red")
title(axes[1],"plot_wireframe")

cs = axes[2].contour(X,Y,Z,zdir ='z',offset=1,norm=norm,cmap=mpl.cm.magma)
cs = axes[2].contour(X,Y,Z,zdir = 'y',offset=2,norm=norm,cmap = mpl.cm.magma)

title(axes[2],"contour")

plt.show()
```
