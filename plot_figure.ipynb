**PLOT 1**

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
import sympy

fig = plt.figure(figsize=(40,20),facecolor="lightcyan")
left,bottom,width,height = 2,2,0.1,0.1
ax = fig.add_axes((left,bottom,width,height),facecolor="#f2f3f2")
x = np.linspace(-2,2,50)
y1 = np.cos(10*x)
y2 = np.exp(-x**2)

ax.plot(x,y1*y2)
ax.plot(x,y2,'b')
ax.plot(x,-y2,'b')
ax.set_xlabel("x")
ax.set_ylabel("y")
```

**PLOT 2**

```python
fig = plt.figure(figsize = (10,10),facecolor="white")
left,bottom,width,height = 1,1,2,2
ax = fig.add_axes((left,bottom,width,height),facecolor= "green")

x = np.linspace(-2,2,100)
y1 = np.sin(10*x)+np.cos(10*x)
ax.plot(x,y1,'white')
ax.set_xlabel("x")
ax.set_ylabel("y")
```

**PLOT 3**

```python
sympy_x = sympy.Symbol("x")
x = np.linspace(-10*np.pi,10*np.pi,100)
def expression(x,n):
  return sympy.lambdify(sympy_x, sympy.cos(2*sympy_x).series(n).removeO(),"numpy")(x)

fig, ax = plt.subplots()
ax.plot(x,np.cos(x),linewidth=2,color="green",label='exact')

colors = ["green","blue"]
linestyles = [":","dashed",'--']
for idx, n in enumerate(range(1,10,2)):
  ax.plot(x,expression(x,n), color = colors[idx%2],linestyle = linestyles[idx//2],linewidth=2)

ax.set_ylim(-1,1)
ax.set_xlim(-1.5*np.pi,1.5*np.pi)
fig.subplots_adjust(right=0.90)
```

**PLOT 4**

```python
x = np.linspace(0,100,100)
y = np.cos(2*x)+np.sin(2*x)
fig,ax = plt.subplots(figsize = (10,10),subplot_kw = {'facecolor':"lightcyan"})
ax.plot(x,y,lw=2)
ax.set_xlabel("x",labelpad=10,fontsize=20,fontname = "Times New Roman",color="red")
ax.set_ylabel("f(x)",labelpad=10,fontsize=20,fontname="Times New Roman",color="red")
```
